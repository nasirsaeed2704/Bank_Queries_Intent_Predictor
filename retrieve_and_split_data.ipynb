```python
import pandas as pd
from sklearn.model_selection import train_test_split

train_val = pd.read_csv('https://raw.githubusercontent.com/PolyAI-LDN/task-specific-datasets/refs/heads/master/banking_data/train.csv')
test_df = pd.read_csv('https://raw.githubusercontent.com/PolyAI-LDN/task-specific-datasets/refs/heads/master/banking_data/test.csv')

train_val['category'] = train_val['category'].map(intent_to_category)
test_df['category'] = test_df['category'].map(intent_to_category)
train_val['category'] = train_val['category'].map(label2id)
test_df['category'] = test_df['category'].map(label2id)

train_df, valid_df = train_test_split(
    train_val,
    test_size=0.2,
    stratify=train_val['category'],
    shuffle=True,
    random_state=42
)

# Check class distribution
for i in train_df['category'].unique():
  print(f"{i}: {round(len(train_df[train_df['category']==i]) / len(train_df['category']), 5)*100}%")
  print(f"{i}: {round(len(valid_df[valid_df['category']==i]) / len(valid_df['category']), 5)*100}%")
  print()

train_df.to_csv('drive/MyDrive/ML/train.csv', index=False)
valid_df.to_csv('drive/MyDrive/ML/valid.csv', index=False)
test_df.to_csv('drive/MyDrive/ML/test.csv', index=False)
```
